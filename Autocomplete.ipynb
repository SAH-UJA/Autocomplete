```
import nltk
nltk.download('brown')
```

```
from nltk.corpus import brown
```

```
corp = brown.words()[:10000]
corp2 = brown.words()[:10000]
```

```
word = {}
inv_word = {}
i = 0
for wrd in set(corp):
  if wrd.lower() not in word.keys():
    word[wrd.lower()] = i
    inv_word[i] = wrd.lower()
    i+=1
corp = word.keys()
```

```
word
```

```
x = []
for i in corp2:
  x.append(word[i.lower()])
```

```
x
```

```
window = 5
xnew = []
ynew = []
for i in range(window,len(x)):
  xnew.append(x[i-window:i])
  ynew.append(x[i])
```

```
xnew
```

```
ynew
```

```
import numpy as np
xn = np.array(xnew)
yn = np.array(ynew)
```

```
import numpy as np
y = np.zeros((len(yn),len(set(corp))))
```

```
yn
```

```
for i, val in enumerate(yn):
  y[i][val] = 1
```

```
import tensorflow as tf
from tensorflow import keras
from keras import layers, models
```

```
model = models.Sequential()
model.add(layers.Input(shape=(5,)))
model.add(layers.Embedding(len(corp),100, input_length=5))
model.add(layers.Bidirectional(layers.GRU(32, return_sequences=True)))
model.add(layers.Bidirectional(layers.GRU(32)))
model.add(layers.Dense(len(set(corp)),activation='softmax'))
```

```
model.summary()
```

```
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```
xn.shape
```

```
history = model.fit(xn,y,validation_split=0.2,batch_size=32,epochs=100)
```

```
sent = ['He','is','not','big','enough']
input_vec = []
for i in sent:
  input_vec.append(word[i.lower()])
```

```
x_test = np.array(input_vec).reshape(1,5)
```

```
next_word = inv_word[np.argmax(model.predict(x_test))]
```
